```python
from gensim.models import KeyedVectors
from gensim.test.utils import datapath

import pandas as pd
import numpy as np
import tqdm
```

```python
w2v_mod = KeyedVectors.load_word2vec_format('your-vector-file-debiased.kv', binary=False)
```

```python
df = pd.read_csv("tagged_tweets.csv")
```

```python
df.hashtags = df.hashtags.apply(eval)
df.head(20)
```

```python
w2v_mod['PokemonLEGENDS'.lower()]
```

```python
def cos_sim(predict, actual):
    result = 0
    mag_pre = 1
    mag_act = 1
    for token in actual:
        if predict[0] == token:
            result = 1
        else:
            try:
                predict_0 = w2v_mod[predict[0]]
                mag_pre = np.linalg.norm(predict_0)
            except:
                predict_0 = np.zeros(100)
                mag_pre = 1
            try:
                actual_i = w2v_mod[token]
                mag_act = np.linalg.norm(actual_i)
            except:
                actual_i = np.zeros(100)
                mag_act = 1
            result = max(result, (sum(np.multiply(predict_0, actual_i)) / mag_pre / mag_act))

    return result
```

```python
cos_sim(['student'], ['class', 'teacher'])
```

```python
df['hashtags_predict'] = df['hashtags']
df['accuracy'] = df.hashtags
```

```python
for index, row in df.iterrows():
    df.accuracy.iloc[index] = cos_sim(df.hashtags_predict.iloc[index], df.hashtags.iloc[index])
    if index % 2000 == 0:
        print("Finished ", index, " predictions.")
print(df.accuracy.mean())
df.head()
```
